```r
exp<-read.csv("GSE16011+GSE61374.csv",header=TRUE,sep=",",quote="\"");
```

```r
mu<-exp[,-1];
a<-mu[,2];
mu<-mu[,-2];
mu
```

```r
mu$class<- NULL
```

```r
gg<- kmeans(mu, 4)
```

```r
gg
```

```r
table(a, gg$cluster) 
```

```r
plot(mu[c("diagno.Ag.at.sis","survival")], col = gg$cluster)
```

```r
d <- dist(mu, method = "euclidean",diag = FALSE, upper = FALSE, p = 2)
```

```r
d
```

```r
e <-hclust(d, method = "median", members=NULL)
plot(e,hang=-1,type="tirangle")
result=cutree(e,k=3)
table(a,result)
```

```r
e <-hclust(d, method = "complete", members=NULL)
plot(e,hang=-1,type="tirangle")
result=cutree(e,k=3)
table(a,result)
```

```r
e <-hclust(d, method = "ward", members=NULL)
plot(e,hang=-1,type="tirangle")
result=cutree(e,k=3)
table(a,result)
```

```r
e <-hclust(d, method = "average", members=NULL)
plot(e,hang=-1,type="tirangle")
result=cutree(e,k=3)
table(a,result)
```
